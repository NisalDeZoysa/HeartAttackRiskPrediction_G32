```python
# Importing necessary libraries for data visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Setting the figure size for the plot
plt.figure(figsize=(8, 7))

# Looping through each column in the 'ctg_data' DataFrame
for i in ctg_data.columns:
    # Counting the occurrences of each unique value in the column
    ctg_num = ctg_data[i].value_counts()
    # Creating a bar plot using seaborn
    chart = sns.barplot(x = ctg_num.index, y = ctg_num)
    for p in chart.patches:
        chart.annotate(format(p.get_height(), '.0f'), 
                     (p.get_x() + p.get_width() / 2., p.get_height()), 
                     ha = 'center', va = 'center', 
                     xytext = (0, 10), 
                     textcoords = 'offset points',
                     fontsize=10)
    # Rotating x-axis labels for better readability
    chart.set_xticklabels(chart.get_xticklabels(), rotation=90)
    # Displaying the plot for the current column
    plt.show()
```

```python
# Splitting the 'Blood Pressure' column into two separate columns: 'BP_Systolic' and 'BP_Diastolic'
data_new[['BP_Systolic', 'BP_Diastolic']] = data_new['Blood Pressure'].str.split('/', expand=True)

# Converting the new columns to numeric values, removing non-numeric characters
# The 'errors='coerce'' argument is used to replace any non-numeric values with NaN
data_new['BP_Systolic'] = pd.to_numeric(data_new['BP_Systolic'], errors='coerce')
data_new['BP_Diastolic'] = pd.to_numeric(data_new['BP_Diastolic'], errors='coerce')

# Displaying the updated DataFrame with the new columns
data_new.head()
```

```python
data_new = data_new.drop('Blood Pressure', axis=1) # Dropping the 'Blood Pressure' column from the DataFrame
data_new = data_new.drop('Patient ID', axis=1) # Dropping the 'Patient ID' column from the DataFrame
data_new.head()
```

```python
data_new["Sex"] = data_new["Sex"].astype('category') # Converting the 'Sex' column to the 'category' data type
data_new["Diet"] = data_new["Diet"].astype('category') # Converting the 'Diet' column to the 'category' data type
data_new["Country"] = data_new["Country"].astype('category') # Converting the 'Country' column to the 'category' data type
data_new["Continent"] = data_new["Continent"].astype('category') # Converting the 'Continent' column to the 'category' data type
data_new["Hemisphere"] = data_new["Hemisphere"].astype('category') # Converting the 'Hemisphere' column to the 'category' data type
```

```python
# Specify the columns you want to include in the new variable
selected_columns = ['Sex', 'Diet', 'Country', 'Continent', 'Hemisphere']

# Create a new variable 'categorical_data' with the selected columns without modifying the original data
categorical_data = data_new[selected_columns].copy()

categorical_data.head()
```

```python
categorical_data["Sex"] = categorical_data["Sex"].cat.codes # Converting the 'Sex' column to numerical codes
categorical_data["Diet"] = categorical_data["Diet"].cat.codes # Converting the 'Diet' column to numerical codes
categorical_data["Country"] = categorical_data["Country"].cat.codes # Converting the 'Country' column to numerical codes
categorical_data["Continent"] = categorical_data["Continent"].cat.codes # Converting the 'Continent' column to numerical codes
categorical_data["Hemisphere"] = categorical_data["Hemisphere"].cat.codes # Converting the 'Hemisphere' column to numerical codes

# Displaying the updated DataFrame after converting categorical columns to numerical codes
categorical_data.head()
```

```python
# Select numerical columns
numerical_data = data_new.select_dtypes(include=['int64', 'float64']).copy()

# Display the numerical data
numerical_data.head()
```

```python
# Combining the categorical and numerical DataFrames using the 'join' method
dataset_new=categorical_data.join(numerical_data)
dataset_new
```

```python
print(dataset_new.dtypes)
```

```python
import numpy as np

dataset = dataset_new.values # Converting the combined DataFrame 'dataset_new' to a NumPy array

# Extracting features (data_1) and target variable (target_1) from the dataset
data_1 = dataset[:, :23]
target_1 = dataset[:, 23]

# Extracting additional columns 'data_2' and 'data_3' from the dataset
data_2 = dataset[:, 24]
data_3 = dataset[:, 25]

# If you want to keep data_2 and data_3 as separate columns in a 2D array
data_2 = data_2.reshape(-1, 1)
data_3 = data_3.reshape(-1, 1)

# concatenate data_2 and data_3 horizontally to data_1
data_1 = np.concatenate((data_1, data_2, data_3), axis=1)
data_1
```

```python
from sklearn.model_selection import train_test_split

# Splitting the data into training and testing sets
# - 'data_1' is the feature array
# - 'target_1' is the target variable
# - 'test_size=0.3' specifies that 30% of the data will be used for testing
# - 'random_state=0' sets a seed for reproducibility
# - 'stratify=target_1' ensures that the target variable's class distribution is preserved in the splits

train_data,test_data,train_target,test_target=train_test_split(data_1,target_1,test_size=0.3, random_state = 0,stratify=target_1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_score

from sklearn.svm import SVC
from sklearn.metrics import precision_score
```

```python
from sklearn.svm import SVC

# Creating a Support Vector Machine (SVM) classifier with an RBF kernel
model=SVC(kernel = 'rbf')
```

```python
model.fit(train_data,train_target)
```

```python
# Making predictions on the test data using the trained Support Vector Machine (SVM) model
predicted_target=model.predict(test_data)

# Importing the accuracy_score function from scikit-learn to evaluate model performance
from sklearn.metrics import accuracy_score

# Calculating the accuracy score by comparing predicted labels with actual labels
acc=accuracy_score(test_target,predicted_target)

print('Accuracy:',acc)
```

```python
from sklearn.metrics import precision_score

# Calculating precision by comparing predicted labels with actual labels
# The 'zero_division' parameter is set to 1 to handle the case where there are no true positive predictions
precision = precision_score(test_target,predicted_target,zero_division=1)

print('Precision of ',precision)
```
